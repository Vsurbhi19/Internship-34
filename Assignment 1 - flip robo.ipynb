# 3. Python program to display IMDB’s Top rated 100 Indian movies

```python
from bs4 import BeautifulSoup
import requests

page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
page

soup= BeautifulSoup(page.content)
soup

first_title = soup.find('h1',class_="header")
first_title.text
```

```python
name=[]
for i in soup.find_all('td',class_="titleColumn"):
    name.append(i.text.split('|') [0])
    
    
    
name
```

```python
rating=[]
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text)
    
    
    
rating
```

```python
yearofrelease=[]
for i in soup.find_all('span',class_="secondaryInfo"):
    yearofrelease.append(i.text)
    
    
    
yearofrelease
```

```python
print(len(name),len(rating),len(yearofrelease))
```

```python
import pandas as pd
df = pd.DataFrame ({'name': name, 'rating':rating, 'yearofrelease': yearofrelease})
df
```

# 9. Python program to scrape mentioned details from dineout.co.in 

```python
from bs4 import BeautifulSoup
import requests

page=requests.get('https://www.dineout.co.in/delhi-restaurants?search_str=restaurants%20near%20CP')
page

soup= BeautifulSoup(page.content)
soup

first_title = soup.find('h1',class_="listing-title")
first_title.text
```

```python
restaurant=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    restaurant.append(i.text)
    
    
    
restaurant
```

```python
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
    
    
location
```

```python
rating=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
    
    
    
rating
```

```python
cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
    
    
    
cuisine
```

```python
imageurl=[]
for i in soup.find_all('img',class_="no-img"):
    imageurl.append(i.get('data-src'))
    
    
    
imageurl
```
